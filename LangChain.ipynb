```python
# 1. Install the latest LangChain-Google integration package
!pip install -qU langchain-google-genai
import os
from google.colab import userdata
# 2. Securely load the API key from Colab Secrets
os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')
print("🚀 Setup Successful! Connected to LangChain.")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
# 1. Choose the current production brain (Upgraded to 2.5)
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
# 2. Give it a Permanent Job/Role
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant. Translate the user's text into French."),
    ("user", "{text}")
])
# 3. Create the Assembly Line (The Chain!)
chain = prompt | model
print("⛓️ Chain built successfully with Gemini 2.5 Flash!")
```

```python
try:
    # 4. Push your sentence through the updated assembly line
    result = chain.invoke({"text": "Good morning class, let's learn AI today!"})
    print("🤖 AI Response:")
    print(result.content)
except Exception as e:
    print(f"❌ Error: {e}")
```

```python
# Cell 4: The Vocabulary Extractor Command
from langchain_core.output_parsers import StrOutputParser
# 1. This command asks the AI to isolate key vocabulary words from the text
vocab_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a French professor. Extract 3 key vocabulary words from the English text provided, translate them to French, and provide a 1-sentence grammatical tip for each."),
    ("user", "{text}")
])
# 2. Build the new command pipeline
vocab_chain = vocab_prompt | model | StrOutputParser()
print("🔤 Extracting French vocabulary and tips...")
classroom_text = "Good morning class, let's learn AI today!"
vocab_results = vocab_chain.invoke({"text": classroom_text})
print(vocab_results)
```

```python
# Cell 5: The French Comprehension Quiz Command
# This command turns the AI into a quiz generator based on the previous translation
quiz_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a French language evaluator. Based on the translated French text provided, generate 2 multiple-choice comprehension questions in French to test a student's understanding. Include an answer key at the bottom."),
    ("user", "Translated French Text: {french_translation}")
])
quiz_chain = quiz_prompt | model | StrOutputParser()
# Let's grab the output from our very first chain (Cell 3)
# Original translation was: "Bonjour la classe, apprenons l'IA aujourd'hui !"
french_text = "Bonjour la classe, apprenons l'IA aujourd'hui !"
print("📝 Generating French Concept Questions...")
quiz_results = quiz_chain.invoke({"french_translation": french_text})
print(quiz_results)
```
